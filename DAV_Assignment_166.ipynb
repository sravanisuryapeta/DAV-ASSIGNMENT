# **Heart Disease**

Heart Disease Data for Health Research

**About Dataset**

This dataset contains various health indicators and risk factors related to heart disease. Parameters such as age, gender, blood pressure, cholesterol levels, smoking habits, and exercise patterns have been collected to analyze heart disease risk and contribute to health research. The dataset can be used by healthcare professionals, researchers, and data analysts to examine trends related to heart disease, identify risk factors, and perform various health-related analyses.

Columns ;

Age: The individual's age.

Gender: The individual's gender (Male or Female).

Blood Pressure: The individual's blood pressure (systolic).

Cholesterol Level: The individual's total cholesterol level.

Exercise Habits: The individual's exercise habits (Low, Medium, High).

Smoking: Whether the individual smokes or not (Yes or No).

Family Heart Disease: Whether there is a family history of heart disease (Yes or No).

Diabetes: Whether the individual has diabetes (Yes or No).

BMI: The individual's body mass index.

High Blood Pressure: Whether the individual has high blood pressure (Yes or No).

Low HDL Cholesterol: Whether the individual has low HDL cholesterol (Yes or No).

High LDL Cholesterol: Whether the individual has high LDL cholesterol (Yes or No).

Alcohol Consumption: The individual's alcohol consumption level (None, Low, Medium, High).

Stress Level: The individual's stress level (Low, Medium, High).

Sleep Hours: The number of hours the individual sleeps.

Sugar Consumption: The individual's sugar consumption level (Low, Medium, High).

Triglyceride Level: The individual's triglyceride level.

Fasting Blood Sugar: The individual's fasting blood sugar level.

CRP Level: The C-reactive protein level (a marker of inflammation).

Homocysteine Level: The individual's homocysteine level (an amino acid that affects blood vessel health).

Heart Disease Status: The individual's heart disease status (Yes or No).

```python
import pandas as pd

# Assuming the file is named 'heart_disease.csv'
df = pd.read_csv('heart_disease.csv')

# Display first few rows
df.head()
```

```python
# Basic Information
# Shape of the dataset
print(df.shape)

# Column names
print(df.columns)

# Data types and non-null counts
print(df.info())

# Summary statistics for numerical columns
print(df.describe())
```

```python
#Accessing Data
# Access a single column
print(df['Age'])

# Access multiple columns
print(df[['Age', 'Gender', 'Exercise Habits']])

# Access rows using iloc (index-based)
print(df.iloc[0])  # First row

# Access rows using loc (label-based)
print(df.loc[0])  # Row with index label 0
```

```python
print(df.columns.tolist())
```

```python
#Filtering Data
# People with cholesterol > 200
print(df[df['Cholesterol Level'] > 200])

# Smokers
print(df[df['Smoking'] == 'Yes'])

# Diabetics with heart disease
print(df[(df['Diabetes'] == 'Yes') & (df['Heart Disease Status'] == 'Yes')])
```

```python
#GroupBy and Aggregation
# Average BMI by gender
print(df.groupby('Gender')['BMI'].mean())

# Count of heart disease cases
print(df['Heart Disease Status'].value_counts())
```

```python
#Sorting
# Sort by age
print(df.sort_values('Age'))

# Sort by cholesterol level descending
print(df.sort_values('Cholesterol Level', ascending=False))
```

```python
#Adding and Removing Columns
# Add age in months
df['age_in_months'] = df['Age'] * 12

# Drop a column
df = df.drop('age_in_months', axis=1)
```

```python
#Missing Values
# Check for missing values
print(df.isnull().sum())

# Fill missing BMI with mean
df['BMI'] = df['BMI'].fillna(df['BMI'].mean())

# Drop all rows with any missing values
df = df.dropna()
```

```python
# Broadcasting in pandas:
# Add 5 years to everyone's age (broadcasting +5)
df['Age_plus_5'] = df['Age'] + 5

# Multiply all cholesterol values by 2
df['Cholesterol_Doubled'] = df['Cholesterol Level'] * 2
print(df[['Age', 'Cholesterol Level', 'Age_plus_5', 'Cholesterol_Doubled']].head())
```

```python
# Difference between systolic BP and cholesterol
df['bp_minus_chol'] = df['Blood Pressure'] - df['Cholesterol Level']

# Sum of BMI and Triglycerides
df['bmi_plus_trig'] = df['BMI'] + df['Triglyceride Level']
# Display first 5 rows of the relevant columns
print(df[['Blood Pressure', 'Cholesterol Level', 'bp_minus_chol',
          'BMI', 'Triglyceride Level', 'bmi_plus_trig']].head())
```

```python
#Conditional Broadcasting
# Create a new column: True if age > 60
df['is_senior'] = df['Age'] > 60

# Create a new column: True if cholesterol level > 240
df['cholesterol_status'] = df['Cholesterol Level'] > 240

# Display first 5 rows to check new columns
print(df[['Age', 'is_senior', 'Cholesterol Level', 'cholesterol_status']].head())
```

```python
#NUMPY OPERATIONS
```

```python
import numpy as np

# Extract columns as NumPy arrays
age = df['Age'].values
chol = df['Cholesterol Level'].values
bmi = df['BMI'].values
trig = df['Triglyceride Level'].values
print("Age:", age[:5])
print("Cholesterol:", chol[:5])
print("BMI:", bmi[:5])
print("Triglycerides:", trig[:5])
```

```python
#Arithmetic Operations
# ✅ 1. Add 5 years to age
age_plus_5 = age + 5

# ✅ 2. Subtract triglyceride from cholesterol
chol_minus_trig = chol - trig

# ✅ 3. Multiply BMI by 2
bmi_double = bmi * 2

# ✅ 4. Print first 5 results
print("Age + 5:", age_plus_5[:5])
print("Cholesterol - Triglyceride:", chol_minus_trig[:5])
print("BMI x 2:", bmi_double[:5])
```

```python
#Mathematical Functions
# Square root of cholesterol
sqrt_chol = np.sqrt(chol)

# Natural log of BMI (add 1 to avoid log(0))
log_bmi = np.log(bmi + 1)

# Triglyceride raised to power 2
trig_squared = np.power(trig, 2)

# Print first 5
print("Sqrt(Cholesterol):", sqrt_chol[:5])
print("Log(BMI):", log_bmi[:5])
print("Triglyceride^2:", trig_squared[:5])
```

```python
#Aggregate Functions
# Mean, median, std
print("Average Age:", np.mean(age))
print("Median BMI:", np.median(bmi))
print("Standard Deviation of Cholesterol:", np.std(chol))
```

```python
#Logical / Boolean Indexing
# Seniors (age > 60)
is_senior = age > 60

# High cholesterol (>= 240)
high_chol = chol >= 240

# Print count of seniors and high cholesterol cases
print("Number of seniors:", np.sum(is_senior))
print("Number of high cholesterol cases:", np.sum(high_chol))
```

```python
# Combine Conditions
# People who are seniors and have high cholesterol
senior_high_chol = np.logical_and(is_senior, high_chol)

print("Seniors with high cholesterol:", np.sum(senior_high_chol))
```

```python
#Matrix Operations
# Get just the 3 columns as a matrix
health_factors = df[['BMI', 'Cholesterol Level', 'Triglyceride Level']].values

# Define weights as a column vector
weights = np.array([0.3, 0.5, 0.2])

# Matrix multiplication to get weighted score
weighted_score = np.dot(health_factors, weights)

# Add result as a new column
df['health_score'] = weighted_score

# View first 5
print(df[['BMI', 'Cholesterol Level', 'Triglyceride Level', 'health_score']].head())
```

```python
#Filter Rows with NumPy Conditions
# Get rows where cholesterol > 240 and BMI > 25
condition = (df['Cholesterol Level'] > 240) & (df['BMI'] > 25)

# Filter using condition
filtered_data = df[condition]

# Show count and sample
print("Filtered rows count:", len(filtered_data))
print(filtered_data[['Age', 'BMI', 'Cholesterol Level']].head())
```

```python
#Reshape & Transpose
# Example: reshape the age array to 2D (required for some ML models)
age_reshaped = age.reshape(-1, 1)  # shape becomes (n, 1)

# Transpose health_factors matrix (flip rows and columns)
transposed = health_factors.T

print("Reshaped age:", age_reshaped.shape)
print("Transposed shape:", transposed.shape)
```

```python
#Trigonometric Operations
import numpy as np
import pandas as pd

# Ensure correct types: Convert to numeric (non-numeric → NaN)
df['Stress Level'] = pd.to_numeric(df['Stress Level'], errors='coerce')
df['Sleep Hours'] = pd.to_numeric(df['Sleep Hours'], errors='coerce')
df['BMI'] = pd.to_numeric(df['BMI'], errors='coerce')

# Extract as NumPy arrays
stress = df['Stress Level'].values
sleep = df['Sleep Hours'].values
bmi = df['BMI'].values

# ✅ 1. Sine of stress levels (radians)
sin_stress = np.sin(np.radians(stress))

# ✅ 2. Cosine of sleep hours (radians)
cos_sleep = np.cos(np.radians(sleep))

# ✅ 3. Tangent of BMI (radians)
tan_bmi = np.tan(np.radians(bmi))

# ✅ 4. Arcsin of normalized stress (must be in [-1, 1])
stress_norm = stress / np.nanmax(stress)  # Normalize between 0–1
arcsin_stress = np.arcsin(stress_norm)

# ✅ Print first 5 results
print("sin(Stress Level):", sin_stress[:5])
print("cos(Sleep Hours):", cos_sleep[:5])
print("tan(BMI):", tan_bmi[:5])
print("arcsin(Normalized Stress):", arcsin_stress[:5])
```
